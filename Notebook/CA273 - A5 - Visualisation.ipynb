### My function for colour and radius

I made a function for use later on when deciding on what colour or radius a particular mark should be.

I took care to make sure the colour are ones that are typically seen as areas being congested and that the double the people meant double the area and not double the radius.

```python
def color_change(c):
    if c == 0:
        return("gray")
    elif (0 < c < 1000):
        return("#93E305")
    elif(1000 <= c < 2000):
        return('#FAEE30')
    elif(2000 <= c < 4000):
        return('#FFBC17')
    elif(4000 <= c < 6000):
        return('#EE7E15')
    elif(8000 <= c < 10000):
        return ("#c24a34")
    else:
        return ("#FF0000")

from math import pi, sqrt
def radius_change(v):
    return sqrt(v/pi * 10)

radius_change(20000)
```

I implemented it that it both saves the images and stores them so it can create the GIF all in one motion without reloading the file.

It turned out after that there was an error with a couple of images, so I ran it for just those images to correct them and then used the code in the next cell to get all the images and create a GIF from them.

```python
data_frames = [df_2008, df_2010, df_2019]
n_set = ["2008", "2010", "2019"]

for i in range(0, len(data_frames)):
    img = []
    for j in range(384, 576):
        img.append(a_frame(j, data_frames[i], n_set[i]))

    img, *imgs = img
    img.save(fp="frames/{:s}.gif".format(n_set[i]), format='GIF', append_images=imgs,
             save_all=True, duration=200, loop=0)
```

```python
for name in n_set:
    # filepaths
    fp_in = "frames/{:s}/frame_*.png".format(name) # Manually put the images in directories by year
    fp_out = "frames/{:s}.gif".format(name)

    img, *imgs = [Image.open(f) for f in sorted(glob.glob(fp_in))]
    img.save(fp=fp_out, format='GIF', append_images=imgs,
             save_all=True, duration=200, loop=0)
```
